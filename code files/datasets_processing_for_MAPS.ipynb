```python
def reverse_geocoding(lat, lon):
    try:
        location = geolocator.reverse(Point(lat, lon))
        return location.raw['display_name']
    except:
        return None
```

```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.point import Point
geolocator = Nominatim()
f=pd.read_excel('nyctrips.xlsx')
g=f.pivot_table(values=['Latitude','Longitude'],index='Path ID',columns='Origin-Destination')
g=g.reset_index()
g.columns= ['_'.join(col) for col in g.columns]
g['address'] = np.vectorize(reverse_geocoding)(g['Latitude_Origin'], g['Longitude_Origin'])
g['destination_address'] = np.vectorize(reverse_geocoding)(g['Latitude_Destination'], g['Longitude_Destination'])
g.to_csv('nyctrips.csv')
g
```
